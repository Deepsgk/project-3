```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import pg_pass
from config import pg_user
import psycopg2

%matplotlib inline
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
# create house df from csv
nobel_csv = "complete.csv"
Nobel_df = pd.read_csv(nobel_csv)
Nobel_df.head()
```

```python
# iterating the columns
for col in Nobel_df.columns:
    print(col)
```

```python
# transform house table
Nobelp_df = ["awardYear", "category", "categoryFullName", "sortOrder","prizeAmount","motivation","award_link","id","name","fullName","gender","laureate_link","birth_date","birth_cityNow","birth_continent","birth_countryNow","birth_locationString"]
Nobelprize1_df = Nobel_df[Nobelp_df].copy()

# check new list
Nobelprize_df.head()
```

```python
# create engine for postgresql
rds_connection_string = (f'{pg_user}:{pg_pass}@localhost:5432/Nobel_prize_db')
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# check table names
engine.table_names()
```

```python
# Write records stored in a DataFrame to a SQL database
Nobelprize1_df.to_sql(name='nobel1_prize', con=engine, if_exists='append', index=False)
```
